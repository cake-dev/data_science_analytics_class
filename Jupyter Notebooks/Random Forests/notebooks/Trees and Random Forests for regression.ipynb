```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Decision trees and Random Forests can also be applied to regression problems, using the `DecisionTreeRegressor` and `RandomForestRegressor` classes.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

## Regression trees

```python
# A 1D regression with decision tree
x = 5 * np.random.rand(100,1)
y = np.sin(x) + 0.1*np.random.randn(100,1) # y = sin(x) + random noise
y[::5] += 3 * (0.5 - np.random.rand(20,1)) # add outliers
y = y.flatten() # set y as a vector

# decision tree
reg_tree = DecisionTreeRegressor(max_depth=5)
reg_tree.fit(x,y)

# plot data
plt.figure(figsize=(12,5))
plt.scatter(x,y,label='data points')

# plot model
x_plot = np.linspace(-1,6,500).reshape(500,1)
y_plot = reg_tree.predict(x_plot)

plt.plot(x_plot,y_plot, c='red',label='ML model')

plt.legend()
```

**Remark 1**: Regression trees are sensitive to outliers

**Remark 2**: tree-based models are very bad at extrapolation

## Random Forest

```python
# decision tree
reg_forest = RandomForestRegressor(max_depth=3)
reg_forest.fit(x,y.flatten())

# plot data
plt.figure(figsize=(12,5))
plt.scatter(x,y,label='data points')

# plot model
x_plot = np.linspace(-1,6,500).reshape(500,1)
y_plot = reg_forest.predict(x_plot)

plt.plot(x_plot,y_plot, c='red',label='ML model')

plt.legend()
```

**Remark 3:** Random Forests are more robust to outliers
