```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us.csv'
cum_data = pd.read_csv(url,index_col='date',parse_dates=True)
cum_data.head()
```

```python
data = cum_data.diff()
data
```

```python
data = data.resample('w').sum()
data.head()
```

```python
fig, ax = plt.subplots(figsize=(15,5))
data.cases.shift(3).plot(ax=ax)

ax2 = ax.twinx()
data.deaths.plot(ax=ax2, c='r')
```
