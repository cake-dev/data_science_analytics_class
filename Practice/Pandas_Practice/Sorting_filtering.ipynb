```python
import pandas as pd
```

```python
# load imdb_1000
imdb_1000 = pd.read_csv('../Data/imdb_1000.csv')
```

```python
imdb_1000.head()
```

```python
# sort the dataframe by star_rating, from lowest to highestFalse
imdb_1000.sort_values(by='star_rating', ascending=False)
```

```python
# sort by 'star_rating' and 'duration'
imdb_1000.sort_values(by=['star_rating', 'duration'], ascending=[False, True])
```

#### Filtering

```python
# example 1: get movies that are at least 200 minutes long
imdb_1000.loc[imdb_1000.duration >= 200] # or imdb_1000[imdb_1000.duration >= 200]
```

```python
# if duration >= 200, replace duration with 'too long'
imdb_1000.loc[imdb_1000.duration >= 200, 'duration'] = 'too long' 
# use loc to select rows by label to be able to assign a value to the selected rows
```

```python
# getting movies with missing 'content_rating'
imdb_1000.loc[imdb_1000.content_rating.isna()]
```

```python
# get all the dramas that are at least 200 minutes long
imdb_1000.loc[(imdb_1000.duration >= 200) & (imdb_1000.genre == 'Drama')]
```

```python
# get all movies with genre: Drame, Crime, Action
imdb_1000.loc[(imdb_1000.genre == 'Drama') | (imdb_1000.genre == 'Crime') | (imdb_1000.genre == 'Action')]
imdb_1000.loc[imdb_1000.genre.isin(['Drama', 'Crime', 'Action'])] # cleaner way (do this when multiple conditions in same column)
```

```python
# get all dramas and crime that are at least 200 minutes long
imdb_1000.loc[imdb_1000.genre.isin(['Drama', 'Crime']) & (imdb_1000.duration >= 200)] # could also do a triple or, but easier to read this way
```
